```python
from data_processing import ExcelReader, MetallurgicalProcess
excel_reader = ExcelReader('datos')
dfs = excel_reader.read_excel_files()
parameters = excel_reader.calculate_parameters()

metallurgical_process = MetallurgicalProcess(
    df=dfs,
    parameters=parameters,
    Raffinate_Density=1.04,
    PLS_Density=0
)
df_final = metallurgical_process.consolidado()
```

```python
df_final = df_final.reset_index(drop=True)
```

```python
selected_columns = ['TCC 2']
```

```python
filtered_df = df_final[df_final['columna'].isin(selected_columns)]
```

```python
# Verificar columnas duplicadas
duplicated_columns = df_final.columns[df_final.columns.duplicated()]
if len(duplicated_columns) > 0:
    print("Columnas duplicadas detectadas:")
    print(duplicated_columns)
else:
    print("No hay columnas duplicadas.")
```

```python
df_final = df_final.loc[:, ~df_final.columns.duplicated()]
```

```python
if df_final.columns.duplicated().any():
    print("Aún hay columnas duplicadas después de los cambios.")
else:
    print("Las columnas ahora son únicas.")
```
